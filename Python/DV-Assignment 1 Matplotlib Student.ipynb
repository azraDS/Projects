## Kütüphaneleri import edin

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import warnings;
warnings.filterwarnings("ignore")
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.arange(0,11,dtype = float)
```

```python
m
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.figure(figsize = (4,3))
plt.plot(m, E, lw = 4, color = "blue")

plt.xlabel("M in g")
plt.ylabel("E in J")
plt.title("E = mc^2")

plt.show()
```

```python
plt.figure(figsize = (4,3))
plt.plot(m, E, lw = 4, color = "orange")

plt.xlabel("M in g")
plt.ylabel("E in J")
plt.title("E = mc^2")

plt.show()
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
plt.figure(figsize = (4,3))

plt.plot(m, E, lw = 2, color = "green")
plt.yscale('log',base=10) 
plt.grid(visible=True, axis ="y", which="both" )

plt.xlabel("M in g")
plt.ylabel("E in J")
plt.title("E = mc^2")
plt.show()
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.plot(labels, july16_2007, color = "b", label =  "July 16 2007")
plt.plot(labels, july16_2020, color = "orange", label =  "July 16 2020")

plt.legend()
plt.show()

```

```python
plt.plot(labels, july16_2007, color = "b", label =  "July 16 2007")
plt.plot(labels, july16_2020, color = "orange", label =  "July 16 2020")


plt.legend(bbox_to_anchor=(1.30, 0.4), loc='lower right')
plt.show()
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
fig, ax = plt.subplots(2, 1, figsize =(10,6))

ax[0].plot(labels, july16_2007, color= "b", linewidth = 2, linestyle = 'solid')
ax[1].plot(labels, july16_2020, color= "b", linewidth = 2, linestyle = 'solid')

ax[0].set_title("july16_2007", fontsize = 8)
ax[1].set_title("july16_2020",fontsize = 8)
plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
fig, ax = plt.subplots(figsize = (8,6))
ax.plot(labels, july16_2007, color = "b", label =  "July 16 2007")
ax.set_ylabel("july 16 2007", color="blue", fontsize=14)

ax2=ax.twinx()
ax2.plot(labels, july16_2020, color = "r", label =  "July 16 2020")
ax2.set_ylabel("july 16 2020", color="red", fontsize= 14)


plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,6))
ax.plot(labels, july16_2007, color = "b", label =  "July 16 2007")
ax.set_ylabel("july 16 2007", color="blue", fontsize=14)
ax.spines["left"].set_visible("False")

ax2=ax.twinx()
ax2.plot(labels, july16_2020, color = "r", label =  "July 16 2020")
ax2.set_ylabel("july 16 2020", color="red", fontsize= 14)
ax2.spines["right"].set_color("red")

plt.show()
```
